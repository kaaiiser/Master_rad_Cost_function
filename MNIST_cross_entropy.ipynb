```python
import tensorflow as tf
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# Preprocess the data
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
# Flatten the input images
X_train = X_train.reshape((-1, 784))
X_test = X_test.reshape((-1, 784))
```

```python
# Convert the labels to one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
# Split the training data into a training set and a validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
```

```python
# Define the model architecture
model_sgd = tf.keras.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(784,)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

model_momentum = tf.keras.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(784,)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

model_adagrad = tf.keras.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(784,)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

model_adam = tf.keras.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(784,)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# Define the optimizers
sgd = tf.keras.optimizers.SGD(learning_rate=0.01)
momentum = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
adagrad = tf.keras.optimizers.Adagrad(learning_rate=0.01)
adam = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# Compile the model with a categorical cross-entropy loss function
model_sgd.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Train the model with mini-batch gradient descent
history_sgd = model_sgd.fit(X_train, y_train, batch_size=64, epochs=30, verbose= True, validation_data=(X_val, y_val))
```

```python
# Compile the model with a categorical cross-entropy loss function
model_momentum.compile(optimizer=momentum,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Train the model with mini-batch gradient descent
history_momentum = model_momentum.fit(X_train, y_train, batch_size=64, epochs=30, verbose= True, validation_data=(X_val, y_val))
```

```python
# Compile the model with a categorical cross-entropy loss function
model_adagrad.compile(optimizer=adagrad,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Train the model with mini-batch gradient descent
history_adagrad= model_adagrad.fit(X_train, y_train, batch_size=64, epochs=30, verbose= True, validation_data=(X_val, y_val))
```

```python
# Compile the model with a categorical cross-entropy loss function
model_adam.compile(optimizer=adam,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Train the model with mini-batch gradient descent
history_adam = model_adam.fit(X_train, y_train, batch_size=64, epochs=30, verbose= True, validation_data=(X_val, y_val))
```

```python
# Extract the training history
loss_adagrad = history_adagrad.history['loss']
val_loss_adagrad = history_adagrad.history['val_loss']

loss_momentum = history_momentum.history['loss']
val_loss_momentum = history_momentum.history['val_loss']

loss_adam = history_adam.history['loss']
val_loss_adam = history_adam.history['val_loss']

loss_sgd = history_sgd.history['loss']
val_loss_sgd = history_sgd.history['val_loss']
```

```python
# Plot the training history
sns.lineplot(x=range(len(loss_adagrad)), y=loss_adagrad, label='Training loss - Adagrad')
# sns.lineplot(x=range(len(val_loss_adagrad)), y=val_loss_adagrad, label='Validation loss - Adagrad')

sns.lineplot(x=range(len(loss_adam)), y=loss_adam, label='Training loss - Adam')
# sns.lineplot(x=range(len(val_loss_adam)), y=val_loss_adam, label='Validation loss - Adam')

sns.lineplot(x=range(len(loss_momentum)), y=loss_momentum, label='Training loss - Momentum')
# sns.lineplot(x=range(len(val_loss_momentum)), y=val_loss_momentum, label='Validation loss - Momentum')

sns.lineplot(x=range(len(loss_sgd)), y=loss_sgd, label='Training loss - SGD')
# sns.lineplot(x=range(len(val_loss_sgd)), y=val_loss_sgd, label='Validation loss - SGD')
```

```python
# Plot the training history
# sns.lineplot(x=range(len(loss_adagrad)), y=loss_adagrad, label='Training loss - Adagrad')
sns.lineplot(x=range(len(val_loss_adagrad)), y=val_loss_adagrad, label='Validation loss - Adagrad')

# sns.lineplot(x=range(len(loss_adam)), y=loss_adam, label='Training loss - Adam')
sns.lineplot(x=range(len(val_loss_adam)), y=val_loss_adam, label='Validation loss - Adam')

# sns.lineplot(x=range(len(loss_momentum)), y=loss_momentum, label='Training loss - Momentum')
sns.lineplot(x=range(len(val_loss_momentum)), y=val_loss_momentum, label='Validation loss - Momentum')

# sns.lineplot(x=range(len(loss_sgd)), y=loss_sgd, label='Training loss - SGD')
sns.lineplot(x=range(len(val_loss_sgd)), y=val_loss_sgd, label='Validation loss - SGD')
```

```python
# Extract the training history
acc_adagrad = history_adagrad.history['accuracy']
val_acc_adagrad = history_adagrad.history['val_accuracy']

acc_momentum = history_momentum.history['accuracy']
val_acc_momentum = history_momentum.history['val_accuracy']

acc_adam = history_adam.history['accuracy']
val_acc_adam = history_adam.history['val_accuracy']

acc_sgd = history_sgd.history['accuracy']
val_acc_sgd = history_sgd.history['val_accuracy']
```
